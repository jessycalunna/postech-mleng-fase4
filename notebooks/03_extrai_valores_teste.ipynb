```python
import json
import yfinance as yf

ticker = yf.Ticker("PETR4.SA")
hist = ticker.history(period="3mo")

closes = [round(float(p), 2) for p in hist["Close"].tail(60).tolist()]

print(json.dumps({"prices": closes}, indent=2))
```
